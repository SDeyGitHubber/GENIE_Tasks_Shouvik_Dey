```python
import os
import numpy as np
import torch
from torchvision import transforms
from PIL import Image
from diffusers import VQModel

import torch
import h5py
import numpy as np
import os

def preprocess_and_encode_hdf5(hdf5_file, model, batch_size=2000):
    """
    Preprocess and encode images from the HDF5 file into latent representations.

    Args:
        hdf5_file (str): Path to the HDF5 file.
        model (LDMPipeline): LDM model for encoding.
        batch_size (int): Batch size for encoding.

    Returns:
        np.ndarray: Array of latent representations.
    """
    latents = []

    # âœ… Open HDF5 file
    with h5py.File(hdf5_file, "r") as f:
        images = f['X_jets']  # Only extract the images
        num_images = images.shape[0]

        print(f"Total images: {num_images}")

        # âœ… Process in batches
        for i in range(0, num_images, batch_size):
            batch = images[i:i + batch_size]  # Load batch
            print(f"Processing batch {i // batch_size + 1}/{(num_images // batch_size) + 1} with shape: {batch.shape}")

            batch_latents = []

            for img in batch:
                # âœ… Convert to PyTorch tensor
                img_tensor = torch.from_numpy(img).float()

                # âœ… Normalize from [0, 255] to [-1, 1]
                img_tensor = img_tensor / 255.0
                img_tensor = (img_tensor - 0.5) / 0.5

                # âœ… Ensure shape is (3, 125, 125)
                if img_tensor.shape == (125, 125, 3):
                    img_tensor = img_tensor.permute(2, 0, 1)

                # âœ… Add batch dimension and move to GPU
                img_tensor = img_tensor.unsqueeze(0).to("cuda")

                # âœ… Encode using VQVAE to latent space (CORRECTED!)
                with torch.no_grad():
                    latent = model.vqvae.encode(img_tensor).latents.cpu().numpy()  # ğŸ”¥ Use .latents
                    batch_latents.append(latent)

            # âœ… Append batch latents
            latents.extend(batch_latents)

    return np.array(latents)

# âœ… Define paths
hdf5_file = "/kaggle/input/dataset-hdf5/quark-gluon_data-set_n139306.hdf5"
latent_save_dir = "/kaggle/working/ldm_latents"

# âœ… Create save directory
os.makedirs(latent_save_dir, exist_ok=True)

# âœ… Preprocess and encode
latents = preprocess_and_encode_hdf5(hdf5_file, model, batch_size=2000)

# âœ… Save the latent representations
latent_file = os.path.join(latent_save_dir, "latents.npy")
np.save(latent_file, latents)

print(f"âœ… Saved latents at {latent_file}")
```

```python
import torch
import numpy as np
from torch.utils.data import DataLoader, TensorDataset
from diffusers import AutoencoderKL
from tqdm import tqdm
import gc

# âœ… Config
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"ğŸ”¥ Using device: {device}")

# âœ… Load raw images into RAM
raw_images = np.load("/content/MyDrive/MyDrive/QuarkGluonBatches/latents.npy").astype(np.float32)
print("ğŸ”¥ Raw image shape:", raw_images.shape)

# âœ… Remove unnecessary dimensions
if raw_images.shape[1] == 1:
    raw_images = raw_images.squeeze(1)  # Remove single dimension â†’ (139306, 3, 31, 31)

# âœ… Normalize images if needed
if raw_images.max() > 1.0:
    raw_images /= 255.0

# âœ… Convert to TensorDataset & DataLoader
images = torch.from_numpy(raw_images).float()
dataset = TensorDataset(images)
BATCH_SIZE = 32  # ğŸ”¥ Smaller batch size to avoid OOM
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False)

# âœ… Load VAE model once (Lazy loading)
print("ğŸš€ Loading VAE model...")
vae = AutoencoderKL.from_pretrained("stabilityai/stable-diffusion-2-1", subfolder="vae").to(device)
vae.eval()

# âœ… Encode the images into latents
latents = []

# âœ… Encoding loop with memory clearing
with torch.no_grad():
    for batch_idx, (batch,) in enumerate(tqdm(loader)):
        batch = batch.to(device)

        # âœ… Resize to (256, 256) if necessary
        if batch.shape[-1] != 256:
            batch = torch.nn.functional.interpolate(batch, size=(256, 256), mode='bilinear', align_corners=False)

        # âœ… Encode batch into latents
        try:
            encoded = vae.encode(batch).latent_dist.sample()
            latents.append(encoded.cpu().numpy())

            # âœ… Clear CUDA memory after each batch
            del batch, encoded
            torch.cuda.empty_cache()
            gc.collect()

        except torch.cuda.OutOfMemoryError:
            print(f"ğŸš« OOM at batch {batch_idx}. Reducing batch size!")
            BATCH_SIZE = max(BATCH_SIZE // 2, 1)  # Reduce batch size by half
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False)

# âœ… Concatenate all latent batches
latents = np.concatenate(latents, axis=0)

# âœ… Save the latents
output_path = "/kaggle/input/encoded-latents/encoded_latents.npy"
np.save(output_path, latents)
print(f"ğŸ”¥ Latents saved as '{output_path}'")
print(f"âœ… Final latent shape: {latents.shape}")
```

```python
import os
import gc
import time
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
from diffusers import UNet2DConditionModel, AutoencoderKL
from torch.cuda.amp import GradScaler, autocast
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm

# âœ… CUDA Environment Fixes
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"  # âœ… More accurate error reporting
torch.cuda.memory._set_allocator_settings("max_split_size_mb:128")  # âœ… Prevent fragmentation

# âœ… Memory Cleanup
torch.cuda.empty_cache()
gc.collect()

# âœ… TF32 Stability Fix
torch.backends.cuda.matmul.allow_tf32 = False
torch.backends.cudnn.allow_tf32 = False

# âœ… Config
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"ğŸš€ Using device: {device}")

# âœ… Memory Management
def free_memory():
    """ğŸ”¥ Free VRAM & clear cache"""
    torch.cuda.empty_cache()
    gc.collect()

free_memory()

# âœ… Load Latents
print("ğŸ”¥ Loading latents...")
latents_path = "/kaggle/input/encoded-latents/encoded_latents.npy"

if not os.path.exists(latents_path):
    print(f"ğŸš« Latent file not found: {latents_path}")
    exit(1)

latents = np.load(latents_path).astype(np.float32)
print(f"âœ… Latents shape: {latents.shape}")

# âœ… Prepare DataLoader with Dynamic Batch Resizing
BATCH_SIZE = 8  # âœ… Static batch size initially
ACCUMULATION_STEPS = 8
SHUFFLE = True

latents_tensor = torch.from_numpy(latents).float().to(device)
dataset = TensorDataset(latents_tensor)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=SHUFFLE, num_workers=0)

# âœ… Load VAE model with AMP & gradient checkpointing
print("ğŸš€ Loading VAE model...")
vae = None
try:
    vae = AutoencoderKL.from_pretrained(
        "stabilityai/stable-diffusion-2-1",
        subfolder="vae",
        force_download=True
    ).to(device)

    # âœ… Half precision for memory efficiency
    vae.half()

    # âœ… Enable gradient checkpointing
    vae.enable_gradient_checkpointing()
    vae.eval()

    print("âœ… VAE model loaded successfully!")

except Exception as e:
    print(f"âŒ VAE model loading failed: {e}")
    exit(1)

# âœ… Load UNet model with OOM Handling
print("ğŸš€ Loading UNet model...")

try:
    free_memory()

    unet = UNet2DConditionModel.from_pretrained(
        "CompVis/stable-diffusion-v1-4",
        subfolder="unet"
    ).to(device)

    # âœ… Half precision for memory efficiency
    unet.half()

    # âœ… Graph Compilation for memory optimization
    unet = torch.compile(unet)

    # âœ… Enable gradient checkpointing
    unet.enable_gradient_checkpointing()

    print("âœ… UNet model loaded successfully!")

except torch.cuda.OutOfMemoryError:
    print("âŒ CUDA OOM during UNet loading. Exiting training.")
    free_memory()
    exit(1)

except Exception as e:
    print(f"âŒ Failed to load UNet: {e}")
    free_memory()
    exit(1)

# âœ… Optimizer, Scaler, and Loss
LEARNING_RATE = 1e-4
EPOCHS = 10
optimizer = optim.AdamW(unet.parameters(), lr=LEARNING_RATE, weight_decay=1e-2)
scaler = GradScaler()
criterion = nn.MSELoss()

# âœ… Dynamic Batch Resizing Function
# âœ… Dynamic Batch Resizing Function
def dynamic_batch_resize(batch_idx, batch):
    """ğŸ”¥ Automatically reduce batch size on OOM"""
    global BATCH_SIZE

    try:
        with autocast():
            # âœ… Decode latent batch back to image with AMP
            with torch.no_grad():
                decoded = vae.decode(batch).sample  # (B, 3, H, W)

                # âœ… Fix channel mismatch by adding a 4th channel
                noise_channel = torch.zeros_like(decoded[:, :1, :, :])  # (B, 1, H, W)
                decoded = torch.cat((decoded, noise_channel), dim=1)  # (B, 4, H, W)

            # âœ… Add noise
            noise = torch.randn_like(decoded)

            # âœ… UNet Forward Pass with Timesteps and Hidden States
            timesteps = torch.randint(0, 1000, (noise.shape[0],), device=device)
            encoder_hidden_states = torch.randn_like(noise)

            output = unet(noise, timesteps, encoder_hidden_states).sample

            # âœ… Loss calculation
            loss = criterion(output, decoded) / ACCUMULATION_STEPS
            scaler.scale(loss).backward()

        return batch  # âœ… Successful batch return

    except torch.cuda.OutOfMemoryError:
        print(f"ğŸš« OOM at batch {batch_idx}. Reducing batch size...")
        
        # âœ… Reduce batch size progressively
        BATCH_SIZE = max(BATCH_SIZE // 2, 1)
        loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=SHUFFLE, num_workers=0)
        
        free_memory()  # âœ… Immediate VRAM cleanup
        return None  # ğŸš« Skip batch on OOM


# âœ… Training Loop with Frequent Memory Cleanup
SAVE_DIR = "/kaggle/working/checkpoints"
os.makedirs(SAVE_DIR, exist_ok=True)

print("ğŸ”¥ Starting training...")

for epoch in range(EPOCHS):
    unet.train()

    running_loss = 0.0
    optimizer.zero_grad()

    for batch_idx, (batch,) in enumerate(tqdm(loader)):
        batch = batch.to(device)

        # âœ… Apply dynamic batch resizing on OOM
        batch = dynamic_batch_resize(batch_idx, batch)
        if batch is None:
            continue  # ğŸš« Skip batch if OOM occurred

        # âœ… AMP Training
        with autocast():
            # âœ… Decode latent batch back to image with AMP
            with torch.no_grad():
                decoded = vae.decode(batch).sample

            # âœ… Add noise
            noise = torch.randn_like(decoded)

            # âœ… Forward pass through UNet
            timesteps = torch.randint(0, 1000, (noise.shape[0],), device=device)
            encoder_hidden_states = torch.randn_like(noise)

            output = unet(noise, timesteps, encoder_hidden_states).sample

            # âœ… Loss calculation
            loss = criterion(output, decoded) / ACCUMULATION_STEPS

        # âœ… Backpropagation with gradient accumulation
        scaler.scale(loss).backward()

        # âœ… Gradient accumulation step
        if (batch_idx + 1) % ACCUMULATION_STEPS == 0 or batch_idx == len(loader) - 1:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()

        running_loss += loss.item() * ACCUMULATION_STEPS

        # âœ… Memory cleanup after every batch
        free_memory()

        if batch_idx % 10 == 0:
            print(f"Epoch [{epoch + 1}/{EPOCHS}] Batch [{batch_idx}/{len(loader)}] Loss: {loss.item():.6f}")

    avg_loss = running_loss / len(loader)
    print(f"ğŸ”¥ Epoch [{epoch + 1}/{EPOCHS}] - Average Loss: {avg_loss:.6f}")

    # âœ… Save model checkpoint
    checkpoint_path = os.path.join(SAVE_DIR, f"unet_epoch_{epoch + 1}.pt")
    torch.save(unet.state_dict(), checkpoint_path)
    print(f"âœ… Model saved at {checkpoint_path}")

print("ğŸ¯ Training completed successfully!")
```

```python
import h5py
import numpy as np
import os
from PIL import Image

# âœ… HDF5 File Path
file_path = '/kaggle/input/dataset-hdf5/quark-gluon_data-set_n139306.hdf5'
output_dir = '/kaggle/working/extracted_images'  # âœ… Output directory for saved images
os.makedirs(output_dir, exist_ok=True)

# âœ… Extract and Save Individual Images
def extract_images(file_path, output_dir, num_images=10):
    """ğŸ”¥ Extract and save images from X_jets."""
    
    with h5py.File(file_path, 'r') as f:
        # âœ… Load the image dataset
        X_jets = f['X_jets'][:]

        print(f"âœ… Extracting {num_images} images...")

        for i in range(min(num_images, X_jets.shape[0])):
            # âœ… Select image (shape: 125x125x3)
            image = X_jets[i]

            # âœ… Normalize the image to [0, 255] and convert to uint8
            img_normalized = ((image - image.min()) / (image.max() - image.min()) * 255).astype(np.uint8)

            # âœ… Convert to PIL image
            img_pil = Image.fromarray(img_normalized)

            # âœ… Save the image
            img_path = os.path.join(output_dir, f"image_{i + 1}.png")
            img_pil.save(img_path)

            print(f"ğŸ”¥ Saved {img_path}")

        print("âœ… Image extraction completed!")

# âœ… Extract 10 sample images
extract_images(file_path, output_dir, num_images=10)
```

```python
import os
import torch
from PIL import Image
from torchvision import transforms
from torch.autograd import no_grad

# âœ… Configurations
device = "cuda" if torch.cuda.is_available() else "cpu"
model_path = "/kaggle/working/checkpoints/ldm_best_model.pt"
input_dir = "/kaggle/input/images"
output_dir = "/kaggle/working/generated_images"

# âœ… Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# âœ… Load LDM Model
print("ğŸš€ Loading LDM model...")
model = torch.load(model_path, map_location=device)
model.eval()
model.to(device)
print("âœ… Model loaded successfully!")

# âœ… Image Transformation Pipeline
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# âœ… Load and process images
image_files = sorted([f for f in os.listdir(input_dir) if f.endswith(".png")])
print(f"ğŸ”¥ Found {len(image_files)} images.")

# âœ… Generate images
with no_grad():
    for idx, image_file in enumerate(image_files, start=1):
        img_path = os.path.join(input_dir, image_file)
        img = Image.open(img_path).convert("RGB")
        img_tensor = transform(img).unsqueeze(0).to(device)

        # âœ… Generate output
        output = model(img_tensor)
        output_img = (output.squeeze(0).cpu().detach().clamp(-1, 1) + 1) / 2  # Denormalize
        output_img = transforms.ToPILImage()(output_img)

        # âœ… Save output image
        output_path = os.path.join(output_dir, f"gen_{idx}.jpg")
        output_img.save(output_path)

        print(f"âœ… Saved: {output_path}")

print("ğŸ¯ All images generated successfully!")
```

```python
import os
import torch
import numpy as np
from PIL import Image
from torchvision import transforms
from lpips import LPIPS
from skimage.metrics import structural_similarity as ssim

# âœ… Configurations
og_dir = "/kaggle/input/images"  # Original images folder
recon_dir = "/kaggle/input/generated-images"  # Reconstructed images folder
common_dim = (256, 256)  # Target common dimension

# âœ… SSIM & LPIPS Metrics Storage
ssim_scores = []
lpips_scores = []

# âœ… Device for LPIPS
device = "cuda" if torch.cuda.is_available() else "cpu"
lpips_model = LPIPS(net='vgg').to(device)

# âœ… Preprocessing Function
def preprocess_image(img_path, target_dim):
    """ğŸ”¥ Preprocess image to the target dimension."""
    img = Image.open(img_path).convert("RGB")
    img = img.resize(target_dim, Image.LANCZOS)

    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    return transform(img)

# âœ… Load and Compare Images
og_images = sorted([f for f in os.listdir(og_dir) if f.endswith(('.png', '.jpg'))])
recon_images = sorted([f for f in os.listdir(recon_dir) if f.endswith(('.png', '.jpg'))])

if len(og_images) != len(recon_images):
    print("âš ï¸ Image count mismatch between original and reconstructed folders!")
    exit(1)

print(f"ğŸ”¥ Found {len(og_images)} image pairs for comparison.")

for og_img, recon_img in zip(og_images, recon_images):
    og_path = os.path.join(og_dir, og_img)
    recon_path = os.path.join(recon_dir, recon_img)

    # âœ… Preprocess both images
    og_tensor = preprocess_image(og_path, common_dim).unsqueeze(0).to(device)
    recon_tensor = preprocess_image(recon_path, common_dim).unsqueeze(0).to(device)

    # âœ… Convert to NumPy for SSIM
    og_np = np.transpose(og_tensor.cpu().numpy().squeeze(), (1, 2, 0))
    recon_np = np.transpose(recon_tensor.cpu().numpy().squeeze(), (1, 2, 0))

    # âœ… Handle SSIM window size dynamically
    min_dim = min(og_np.shape[0], og_np.shape[1])
    win_size = min(7, min_dim if min_dim % 2 != 0 else min_dim - 1)

    # âœ… SSIM Calculation
    ssim_score = ssim(
        og_np, 
        recon_np, 
        channel_axis=2, 
        data_range=2.0, 
        win_size=win_size
    )
    ssim_scores.append(ssim_score)

    # âœ… LPIPS Calculation
    with torch.no_grad():
        lpips_score = lpips_model(og_tensor, recon_tensor).item()
        lpips_scores.append(lpips_score)

    print(f"âœ… Compared: {og_img} & {recon_img} â†’ SSIM: {ssim_score:.4f}, LPIPS: {lpips_score:.4f}")

# âœ… Display Final Scores
print("\nğŸ”¥ **Final Average Scores:**")
print(f"âœ… Average SSIM: {np.mean(ssim_scores):.4f}")
print(f"âœ… Average LPIPS: {np.mean(lpips_scores):.4f}")
```

```python
import os
import torch
import numpy as np
from torchvision import transforms
from PIL import Image
from torchmetrics.image.fid import FrechetInceptionDistance
from torchmetrics.image.inception import InceptionScore

# âœ… Configurations
og_dir = "/kaggle/input/images"  # Original images folder
recon_dir = "/kaggle/input/generated-images"  # Reconstructed images folder
common_dim = (256, 256)  # Target common dimension
device = "cuda" if torch.cuda.is_available() else "cpu"

# âœ… Preprocessing Function
def preprocess_image(img_path, target_dim):
    """ğŸ”¥ Preprocess image and return as uint8 tensor."""
    img = Image.open(img_path).convert("RGB")
    img = img.resize(target_dim, Image.LANCZOS)

    transform = transforms.Compose([
        transforms.ToTensor()  # Keep in [0, 1] range
    ])

    img_tensor = transform(img).unsqueeze(0).to(device)

    # âœ… Convert to uint8 format (0-255 range)
    img_uint8 = (img_tensor * 255).byte()
    
    return img_uint8

# âœ… Load Images into Tensors
def load_images_as_tensor(img_dir, target_dim, device):
    img_files = sorted([f for f in os.listdir(img_dir) if f.endswith(('.png', '.jpg'))])
    images = []

    for img_file in img_files:
        img_path = os.path.join(img_dir, img_file)
        img_tensor = preprocess_image(img_path, target_dim)
        images.append(img_tensor)

    return torch.cat(images)

# âœ… Load images
print("ğŸ”¥ Loading images...")
og_images = load_images_as_tensor(og_dir, common_dim, device)
recon_images = load_images_as_tensor(recon_dir, common_dim, device)

print(f"âœ… Loaded {len(og_images)} original and {len(recon_images)} reconstructed images.")

# âœ… FID Calculation
print("ğŸ”¥ Calculating FID...")
fid = FrechetInceptionDistance(feature=2048).to(device)
fid.update(og_images, real=True)
fid.update(recon_images, real=False)
fid_score = fid.compute().item()
print(f"âœ… FID Score: {fid_score:.4f}")

# âœ… Inception Score Calculation
print("ğŸ”¥ Calculating Inception Score...")
is_model = InceptionScore(feature=2048).to(device)
is_model.update(og_images)
is_model.update(recon_images)
is_mean, is_std = is_model.compute()
print(f"âœ… Inception Score: {is_mean:.4f} Â± {is_std:.4f}")

# âœ… Display Final Results
print("\nğŸ¯ **Final Results:**")
print(f"ğŸ”¥ FID Score: {fid_score:.4f}")
print(f"ğŸ”¥ Inception Score: {is_mean:.4f} Â± {is_std:.4f}")
```
